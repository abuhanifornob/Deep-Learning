**Sigmoid**

```python
import math 
def sigmoid(x):      # Range(0-1)
  return (1/(1+math.exp(-x)))
```

```python
sigmoid(6)
```

```python
sigmoid(-6)
```

```python
sigmoid(4.50)
```

```python
sigmoid(-70)
```

Ta**nh Activation function**

```python
def tanh(x):  # tanh function Range (-1,1)
   return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(20)
```

```python
tanh(-20)
```

```python
tanh(5)
```

```python
tanh(-0.2)
```

```python
tanh(1.5)
```

**Relu Activation Fuction**

```python
def relu(x): # Relu Function Range(0, any Number)
  return max(0,x)
```

```python
relu(10)
```

```python
relu(-10)
```

**Leaky Relu Activation Function**

```python
def leaky_relu(x): # leaky_ Relu function Range is(Nagative any Nuber To Possitive Any Number)
  return max(0.1*x,x)
```

```python
leaky_relu(-10)
```

```python
leaky_relu(25)
```

```python
leaky_relu(-25)
```
