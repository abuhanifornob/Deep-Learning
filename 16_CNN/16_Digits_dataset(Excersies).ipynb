```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras import datasets,models,layers
```

```python
(X_train,y_train),(X_test,y_test)=datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
y_test[:20]
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
ann_model=models.Sequential([
                             layers.Flatten(input_shape=(28,28)),
                             layers.Dense(3000,activation='relu'),
                             layers.Dense(1000,activation='relu'),
                             layers.Dense(10,activation='sigmoid')
])
ann_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
ann_model.fit(X_train,y_train,epochs=5)
```

```python
y_predict=ann_model.predict(X_test)
```

```python
y_predict=[np.argmax(i) for i in y_predict]
y_predict[:5]
```

```python
ann_model.evaluate(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_predict))
```

```python

cnn_model=models.Sequential([
                             layers.Conv2D(filters=50,kernel_size=(3,3),input_shape=(28,28,1),activation='relu'),
                             layers.MaxPooling2D((2,2)),
                             layers.Conv2D(filters=100,kernel_size=(3,3),activation='relu'),
                             layers.MaxPooling2D((2,2)),

                             layers.Flatten(),
                             layers.Dense(100,activation='relu'),
                             layers.Dense(20,activation='relu'),
                             layers.Dense(10,activation='softmax')
                                                      
])
cnn_model.compile(
    optimizer='SGD',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
cnn_model.fit(X_train,y_train,epochs=5)
```
