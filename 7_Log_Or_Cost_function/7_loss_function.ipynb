We are Implement 3 Loss Function-
1. Mean Absolute Error
2. Mean Square Error
3. Log loss/ Binary Cross Entropy


```python
y_predicted = np.array([1,1,0,0,1]) # Prediction Array Create 
y_true = np.array([0.30,0.7,1,0,0.5])#  True Array
```

```python
#........Implement Mean Absolute Error...........................

def mean_absolute_error(y_predicted,y_true):
  total_error=0
  for p,t in zip(y_predicted,y_true): # Zip Function use for 2 value count
    total_error+=abs(p-t)
  mean_error=total_error/len(y_true)
  print('Total Error: ',total_error)
  return mean_error
    
```

```python
mean_absolute_error(y_predicted,y_true)
```

**Mean Squared Absolute Error**

```python
# ....................Implement wothout numpy function
def mean_squared_error(y_predicted,y_true):
  total_error=0
  for yp,yt in zip(y_predicted,y_true):
    total_error+=(yp-yt)**2
  mean_error=total_error/len(y_true)
  print('Total Error: ',total_error)
  return mean_error
```

```python
mean_squared_error(y_predicted,y_true)
```

```python
def binary_cross_entropy(y_predicted,y_true):
  epsilon=1e-15
  y_predicted_new=[max(i,epsilon) for i in y_predicted]
  y_predicted_new=[min(i,1-epsilon) for i in y_predicted_new]
  y_predicted_new=np.array(y_predicted_new)
  return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
binary_cross_entropy(y_predicted,y_true)
```
